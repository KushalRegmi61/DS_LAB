```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Number of samples in the original distribution
num_samples = 10000

# Set the random seed for reporducibility
np.random.seed(100)
```

```python
# Generate samples from a uniform distribution between 0 and 1
population = np.random.uniform(0, 1, num_samples)
```

```python
# Generate samples from an exponential distribution
population = np.random.exponential(scale=2, size=num_samples)
```

```python
# Plot the histogram to visualize the original distribution
plt.hist(population, bins=100, alpha=0.6, color='g')
plt.title('Histogram of the Original Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
sample_size = 100    # Size of each sample
```

```python
sample_means = np.array([])  # To store the means of the samples

# Draw samples and calculate means
for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=False) # Draw a sample
    sample_means = np.append(sample_means, np.mean(sample))                # Calculate the sample mean
```

```python
# Plot the distribution of sample means
plt.figure(figsize=(6, 4))
plt.hist(sample_means, bins=30, color='green', edgecolor='lightgreen', alpha=0.7)
plt.title('Distribution of Sample Means (CLT)', fontsize=11)
plt.xlabel('Sample Mean', fontsize=10)
plt.ylabel('Density', fontsize=10)
plt.grid(alpha=0.3)
plt.show()
```

```python
print("Population Mean = ", population.mean())
print("Population Variance = ", population.var())
```

```python
print("Mean of Sample Means = ", sample_means.mean())
print("Variance of Sample Means = ", sample_means.var())
```
