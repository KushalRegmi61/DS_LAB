# Poisson Distribution
It is a discrete probability distribution that expresses the **probability of a given number of events occuring**, in a **fixed interval of time**, if these **events occur with a known constant mean rate** & **independently of the time since the last event.**

Mathematically, a discrete random variable is said to have a Poisson's distribution, with parameter >0, if it has probability mass function given by:
<br/>f(x; ) = P(X=x) = e^(-)/((^x)*x!)
<br/>Where, <br/>x = no. of successes of some random event.
       <br/> = mean occurence of that random event.

The **hypergeometric distribution** is a probability distribution that describes the likelihood of drawing a certain number of successes in a sample when sampling **without replacement** from a finite population. It is often used in scenarios where we draw a sample from a population and are interested in how many successes (or specific outcomes) we get, where each individual can only be chosen once.

### Hypergeometric Probability Formula:
The probability of getting exactly **k** successes in a sample of size **n** drawn from a population of size **N** with **K** successes is given by the following formula:

\[
P(X = k) = \frac{\binom{K}{k} \binom{N-K}{n-k}}{\binom{N}{n}}
\]

Where:
- \(\binom{K}{k}\) is the number of ways to choose **k** successes from **K** successes.
- \(\binom{N-K}{n-k}\) is the number of ways to choose the remaining **n-k** failures from the **N-K** failures.
- \(\binom{N}{n}\) is the total number of ways to choose **n** items from the entire population of size **N**.



```python
# from scipy.stats import hypergeom


# # Simulate num_samples draws
# samples = hypergeom.rvs(N, K, n, size=num_samples)
# samples = np.random.hypergeometric()

# Parameters for the hypergeometric distribution
N = 52       # Total number of cards in the deck
K = 13       # Number of hearts in the deck
n = 42       # Number of cards drawn
num_samples = 10000  # Number of samples to simulate

# Draw 10,000 samples from the hypergeometric distribution
samples = np.random.hypergeometric(K, N - K, n, num_samples)

# Plotting the histogram using seaborn's histplot
plt.figure(figsize=(8, 6))
sns.histplot(samples, bins=10, kde=False, color='skyblue', edgecolor='black')

# Add titles and labels
plt.title(f'Hypergeometric Distribution (N={N}, K={K}, n={n})')
plt.xlabel('Number of hearts drawn')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Negative Binomial Distribution (Brief Overview):

The **Negative Binomial Distribution** models the number of trials needed to achieve a specified number of successes in a sequence of independent Bernoulli trials. Unlike the binomial distribution, which counts the number of successes in a fixed number of trials, the negative binomial distribution counts the number of trials required to achieve a given number of successes.

### Formula:
The probability mass function (PMF) for the negative binomial distribution is:

\[
P(X = k) = \binom{k + r - 1}{k} p^r (1 - p)^k
\]

Where:
- \( k \) = number of failures before the **r-th** success
- \( r \) = number of successes
- \( p \) = probability of success on each trial
- \( X \) = total number of trials (failures + successes)

### **Normal Distribution: A Brief Introduction**

The **normal distribution** (also known as Gaussian distribution) is a continuous probability distribution that is symmetric around its mean. It is characterized by its bell-shaped curve and is defined by two parameters:
- **Mean ()**: The central value of the distribution (location of the peak).
- **Standard Deviation ()**: A measure of the spread of the distribution.

The **probability density function (PDF)** of the normal distribution is given by the formula:

\[
f(x | \mu, \sigma) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2} \left(\frac{x - \mu}{\sigma}\right)^2}
\]

Where:
- \( x \) is a value from the distribution.
- \( \mu \) is the mean of the distribution.
- \( \sigma \) is the standard deviation.
- \( e \) is Euler's number.



```python


# Parameters for the normal distribution
mu = .42      # mean
sigma = 1   # standard deviation
num_samples = 10000  # number of samples to simulate

# Generate random samples from a normal distribution
samples = np.random.normal(mu, sigma, num_samples)

# Plotting the histogram using seaborn's histplot
plt.figure(figsize=(8, 6))
sns.histplot(samples, bins=30, kde=True, color='skyblue', edgecolor='black')

# Add titles and labels
plt.suptitle("080BCT042 Normal Distribution", fontsize=16)
plt.title(f'Normal Distribution (Mean={mu}, SD={sigma})')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

# Central Limit Theorem (CLT) - Short Introduction
The Central Limit Theorem (CLT) is a fundamental concept in statistics that states that the distribution of the sample mean (or sum) of a large number of independent, identically distributed random variables will approach a normal distribution (bell-shaped curve), regardless of the original distribution of the data.

```python
# Parameters
num_samples = 10000
sample_size = 42
K = 13    # For Hypergeometric Distribution
N = 52    # Total cards
n = 42    # Cards drawn for Hypergeometric

# Creating sample means for different distributions

# Uniform Distribution
UD_sample_means = [np.mean(np.random.uniform(size=sample_size)) for _ in range(num_samples)]

# Binomial Distribution (with p = 0.42, size = 1 for each draw)
BD_sample_means = [np.mean(np.random.binomial(1, 0.42, size=sample_size)) for _ in range(num_samples)]

# Poisson Distribution
PD_sample_means = [np.mean(np.random.poisson(lam=42, size=sample_size)) for _ in range(num_samples)]

# Hypergeometric Distribution
HD_sample_means = [np.mean(np.random.hypergeometric(K, N - K, n, sample_size)) for _ in range(num_samples)]

# Negative Binomial Distribution
NB_sample_means = [np.mean(np.random.negative_binomial(n=5, p=0.3, size=sample_size)) for _ in range(num_samples)]

# Creating subplots for each distribution
plt.suptitle("080BCT042")
fig, ax = plt.subplots(ncols=2, nrows=3, figsize=(12, 12))

# Uniform Distribution plot with seaborn's histplot and kde=True
sns.histplot(UD_sample_means, kde=True, ax=ax[0, 0], bins=30, color='skyblue', edgecolor='black', alpha=0.7)
ax[0, 0].set_title('Sample Means: Uniform Distribution')
ax[0, 0].set_xlabel('Sample Mean')
ax[0, 0].set_ylabel('Frequency')

# Binomial Distribution plot with seaborn's histplot and kde=True
sns.histplot(BD_sample_means, kde=True, ax=ax[0, 1], bins=30, color='lightgreen', edgecolor='black', alpha=0.7)
ax[0, 1].set_title('Sample Means: Binomial Distribution')
ax[0, 1].set_xlabel('Sample Mean')
ax[0, 1].set_ylabel('Frequency')

# Poisson Distribution plot with seaborn's histplot and kde=True
sns.histplot(PD_sample_means, kde=True, ax=ax[1, 0], bins=30, color='lightcoral', edgecolor='black', alpha=0.7)
ax[1, 0].set_title('Sample Means: Poisson Distribution')
ax[1, 0].set_xlabel('Sample Mean')
ax[1, 0].set_ylabel('Frequency')

# Hypergeometric Distribution plot with seaborn's histplot and kde=True
sns.histplot(HD_sample_means, kde=True, ax=ax[1, 1], bins=30, color='gold', edgecolor='black', alpha=0.7)
ax[1, 1].set_title('Sample Means: Hypergeometric Distribution')
ax[1, 1].set_xlabel('Sample Mean')
ax[1, 1].set_ylabel('Frequency')

# Negative Binomial Distribution plot with seaborn's histplot and kde=True
sns.histplot(NB_sample_means, kde=True, ax=ax[2, 0], bins=30, color='purple', edgecolor='black', alpha=0.7)
ax[2, 0].set_title('Sample Means: Negative Binomial Distribution')
ax[2, 0].set_xlabel('Sample Mean')
ax[2, 0].set_ylabel('Frequency')

# Hide the empty subplot at position (2, 1)
ax[2, 1].axis('off')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# function to normalize the columns data
def generate_sample_means(data, sample_size, num_samples):
  generated_samples = np.zeros(num_samples)
  for i in range(num_samples):

    sample = np.random.choice(data, size=sample_size, replace=True, p=None)

    generated_samples[i] = np.mean(sample)

  return generated_samples


# normalize columns
columns = ['math_score', 'reading_score', 'writing_score', 'science_score', 'total_score']

normalized_columns= [generate_sample_means(pd.to_numeric(df[column]), sample_size=42, num_samples=1000) for column in columns]

# math_score column
fig, ax = plt.subplots(ncols=2, nrows=3, figsize=(12, 12))


create_histplot(normalized_columns[0], xaxistitle="Maths_Marks",color= 'blue', ax =ax[0][0])

# reading_score column
create_histplot(normalized_columns[1], xaxistitle="REading_Marks",color= 'blue', ax =ax[0][1])

# writing_score
create_histplot(normalized_columns[2], xaxistitle="writing_score",color= 'blue', ax =ax[1][0])

# Scirence_score
create_histplot(normalized_columns[3], xaxistitle="science_score",color= 'blue', ax =ax[1][1])

create_histplot(normalized_columns[4], xaxistitle="Total_score",color= 'blue', ax =ax[2][0])

# Hide the empty subplot at position (2, 1)
ax[2, 1].axis('off')

# Adjust layout for better spacing
plt.tight_layout()

plt.show()
```
