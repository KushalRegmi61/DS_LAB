#FizBuzz Game

```python
def fizzbuzz(n):
    for i in range(1, n + 1):
        if i % 3 == 0 and i % 5 == 0:
            print("FizzBuzz")
        elif i % 3 == 0:
            print("Fizz")
        elif i % 5 == 0:
            print("Buzz")
        else:
            print(i)

# Play FizzBuzz up to 100
fizzbuzz(100)
```

#Student Details

```python
class Student: # creating a class student
  def __init__(self):
    self.name = input("Enter the name of the student: ")
    self.roll_no = input("Enter the roll number of the student:")

  def display(self):
    print("Student Details")
    print("Name: ", self.name)
    print("Roll No: ", self.roll_no)

# creating an instance of the class
student = Student()
student.display()

```

<h1>Factorial Function</h1>

```python
def factorial(n):
  if n == 0:
    return 1
  else:
    return n * factorial(n-1)

#calculating the factorial of the number
n = int(input("Enter the number: "))
#display
if n >=0:
  print(f"The factorial of the number {n} is {factorial(n)}")
else:
  print("Factorial is not defined for negative numbers")
```

#Fibonacci Number:

```python
def fibonacci_number(n):
  if n == 0:
    return 0
  elif n == 1:
    return 1
  else:
    return fibonacci_number(n-1) + fibonacci_number(n-2)

num = int(input("Enter the number: "))

if num >= 0:
  print(f"The {num} fibonacci number is: {fibonacci_number(num-1)}")
else:
  print("Fibonacci number is not defined for negative numbers")
```

#BMI


```python
class BMI:
  def __init__(self):
    self.weight = float(input("Enter the weight of the person in  Kg: "))
    self.input_height =input('Enter the height of the person.. Eg: 5ft5" : ')
    self.slpitted = self.input_height.split('ft')
    self.feet = float(self.slpitted[0])
    self.inch = float(self.slpitted[1].split('"')[0])
    self.height =  self.feet * 0.3048 + self.inch * 0.0254

  def calculate_bmi(self):
    bmi = self.weight / (self.height ** 2)
    return bmi
# creating a bmi index class
bmi = BMI()

print(f"The BMI index is: {bmi.calculate_bmi()}")
```
