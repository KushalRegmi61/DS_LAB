# Line Plot
For more, see:-
- https://www.datacamp.com/tutorial/matplotlib-time-series-line-plot
- https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html#

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, color='blue', linewidth=2, marker='.', markerfacecolor='red', markeredgecolor='black', markersize=7)
```

```python
plt.plot(x, y, color='blue', linewidth=2)

# Add labels, title, and grid
plt.title('Line Plot')
plt.xlabel('x')
plt.ylabel('y = 4 + sin(0.7x)')
plt.grid(linewidth=0.4, alpha=0.7)

# Show the plot
plt.show()
```

# Heatmap
- https://matplotlib.org/stable/gallery/images_contours_and_fields/image_annotated_heatmap.html
- https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.imshow.html#matplotlib.axes.Axes.imshow

```python
import pandas as pd

# Data preparation

vegetables = ["cucumber", "tomato", "lettuce", "asparagus",
              "potato", "wheat", "barley"]

farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."]

harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])

df = pd.DataFrame(harvest, index=vegetables, columns=farmers)
df
```

```python
# Plotting the heatmap
plt.figure()
plt.imshow(df, cmap='hot_r')
plt.colorbar(label="Harvest Amount")
plt.xticks(ticks=np.arange(len(farmers)), labels=farmers, rotation=45, ha='right')
plt.yticks(ticks=np.arange(len(vegetables)), labels=vegetables)
plt.title("Harvest Data Heatmap")
plt.xlabel("Farmers")
plt.ylabel("Vegetables")
plt.show()
```

# Subplots
**See:**
- https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html

```python
# Define the data
x = np.linspace(0, 10, 100)
y1 = x**2
y2 = x + 4
y3 = np.sqrt(x)
y4 = np.sin(x)
y5 = np.cos(x)
y6 = x**3

# Create a 2x3 grid of subplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12, 8))

# Plot in each subplot
axes[0, 0].plot(x, y1, label='y = x^2')
axes[0, 0].set_title('y = x^2')

axes[0, 1].plot(x, y2, 'r', label='y = x + 4')
axes[0, 1].set_title('y = x + 4')

axes[0, 2].plot(x, y3, 'g', label='y = sqrt(x)')
axes[0, 2].set_title('y = sqrt(x)')

axes[1, 0].plot(x, y4, 'm', label='y = sin(x)')
axes[1, 0].set_title('y = sin(x)')

axes[1, 1].plot(x, y5, 'c', label='y = cos(x)')
axes[1, 1].set_title('y = cos(x)')

axes[1, 2].plot(x, y6, 'k', label='y = x^3')
axes[1, 2].set_title('y = x^3')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

# 3D Surface Plot

```python
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y) # Z is calculated as np.sin(np.sqrt(X**2 + Y**2)).
                         # This requires every possible combination of X and Y values.
Z = np.sin(np.sqrt(X**2 + Y**2))
```

```python
# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, cmap='viridis')

# Add labels and title
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Surface Plot")
ax.view_init(elev=45, azim=120)

plt.show()
```

# Contour Plot

```python
# Create a figure for the contour plots
plt.figure(figsize=(8, 6))

# Plot contour of the loss surface
contour = plt.contourf(X, Y, Z, levels=25, cmap='viridis', alpha=0.8)
plt.colorbar(label='Loss')
plt.xlabel('Weight 1')
plt.ylabel('Weight 2')
plt.title('Loss Function Contour Plot')
```
